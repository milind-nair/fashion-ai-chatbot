```python
urls=[]
name=[]
price=[]
img=[]
brand=[]
features=[]
for j in range(9):
    page = requests.get(df_links[j])
    soup = bs(page.content, 'html.parser')
    print(j)

#it gives us the visual representation of data
    for i in range(40):
        if(j==7):
            div_element = soup.find_all('a', class_='_2rpwqI')[i]
            #x=map(col col-3-12 _2H87wv,col col-9-12 _2vZqPX)
        else:
            div_element = soup.find_all('a', class_='_2UzuFa')[i]
        
        urls.append("https://www.flipkart.com" + str(div_element['href']))
        pageInside = requests.get(urls[len(urls)-1])
        soup1 = bs(pageInside.content, 'html.parser')
        if(j!=7):
            soup2 = soup1.find('div', class_='X3BRps')
            keys = soup2.select('div',class_='.col.col-3-12._2H87wv')
        else:
            keys = soup1.select('td', class_='_1hKmbr col col-3-12')


        k=1    
        dict={}
        print(len(keys))
        while k<len(keys)-1:
            
            key_text = keys[k].get_text().strip()
            value_text = keys[k+1].get_text().strip()
            k=k+3
            dict[key_text]=value_text
            #print(f"Key: {key_text}\nValue: {value_text}\n")

        name_element = soup1.find('span', class_='B_NuCI')
        price_element = soup1.find('div', class_='_30jeq3')
        #descrip_element=soup1.find('div', class_='_30jeq3')
        img_elements = soup1.select('img[class*="_396"]')
        img_element=''
        for image in img_elements:
            img_element = image['src']
            
    
        #_2r_T1I _396QI4
        soup2=bs(str(soup1.find('span',class_='G6XhRU')),'html.parser')
        brand_element=soup2.text.replace('\xa0', '')
        brand.append(brand_element)
        img.append(img_element)
        name.append(name_element.text)
        price.append(price_element.text)
        features.append(dict)
df['pdt_url']=urls
df['pdt_name']=name
df['pdt_categ']="Men Top"
df['price']=price
df['pdt_img']=img
df['brand']=brand
df['Features']=features


#B_NuCI
#_30jeq3 _16Jk6d
```
